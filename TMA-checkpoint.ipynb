```python
#載入套件庫與讀入資料
import pandas as pd
import matplotlib.pyplot as plt
import sqlite3
import numpy as np

db = sqlite3.connect('TWStock_2')
eva = pd.read_sql(con=db,sql='SELECT * FROM "2603"')
```

```python
#清理資料
#我們需要成交股數、開盤價、最高價、最低價、收盤價的資料
#並使用Date當作我們的索引值

eva.index = pd.to_datetime(eva['Date'])
eva = eva[['收盤價']]
eva.columns = ['Close']
eva['Close'] = pd.to_numeric(eva['Close']) #字串轉數值
eva.head()
```

```python
x=20

#我們分別計算20天,60天的移動平均線
eva['MA_20'] = eva['Close'].rolling(round((x+1)/2)).mean()
eva['TMA_20'] = eva['MA_20'].rolling(round((x+1)/2)).mean()

eva.head()
```

```python
eva = eva['2021-01-01':] #2021年開始
```

```python
eva['Signal'] = 0.0
eva['Signal'] = np.where(eva['TMA_20'] < eva['Close'], 1.0, 0.0)
eva['Position'] = eva['Signal'].diff()
```

```python
plt.figure(figsize = (20,10))
#畫線 
eva['Close'].plot(color = 'k') 
eva['TMA_20'].plot(color = 'b',label = '20TMA') 
# 買進信號
plt.plot(eva[eva['Position'] == 1].index, 
         eva['TMA_20'][eva['Position'] == 1], 
         '^', markersize = 15, color = 'r', label = 'buy')
#賣出信號
plt.plot(eva[eva['Position'] == -1].index, 
         eva['TMA_20'][eva['Position'] == -1], 
         'v', markersize = 15, color = 'g', label = 'sell')
plt.ylabel('Price in Rupees', fontsize = 15 )
plt.xlabel('Date', fontsize = 15 )
plt.title('EVA(2603) 20TMA Crossover close', fontsize = 20)
plt.legend()
plt.grid()
plt.show()
```
